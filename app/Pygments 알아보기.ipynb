```python
from pygments.lexers import get_all_lexers
from pygments.styles import get_all_styles
```

```python
LEXERS = []
for item in get_all_lexers():
#     print(item)
    if item[1]:
#         print()
        LEXERS.append(item)
#         print(LEXERS)
```

```python
LANGUAGE_CHOICES = []
for item in LEXERS:
    cur_tuple = (item[1][0], item[1])
    print(cur_tuple)
    LANGUAGE_CHOICES.append(cur_tuple)
#     print(LANGUAGE_CHOICES)
LANGUAGE_CHOICES.sort()
```

```python
STYLE_CHOICE = []
for item in get_all_styles():
    print(item)
    cur_tuple = (item, item)
    STYLE_CHOICE.append(cur_tuple)
STYLE_CHOICE.sort()
```
